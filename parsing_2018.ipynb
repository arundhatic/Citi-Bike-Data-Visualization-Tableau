```python
import pandas as pd
import numpy as np
from geopy import distance
from geopy import Point
```

```python
rawData = pd.read_csv("data/nybike_from_url_2018.csv")
df_18 = pd.DataFrame(rawData)
df_18.head()
```

```python
df_18.tail()
```

```python
df_18['Gender']
```

```python
df_18['Gender'] = df_18['Gender'].replace([0,1,2],['unknown',"Male","Female"])
df_18["Distance"] = ''
df_18.tail()
```

```python
for index, row in df_18.iterrows():
    coords_1 = (row['Start Station Latitude'],row['Start Station Longitude'])
    coords_2 = (row['End Station Latitude'],row['End Station Longitude'])
    df_18.at[index,'Distance'] = distance.distance(coords_1, coords_2).km  
   # df_18_19_20.set_value(index,'distance',distance.vincenty(coords_1, coords_2).km)
df_18.tail()

```

```python
df_18.to_csv('data/2018_parsed_citibikenyc_tripdata.csv', index_label = "TripID")
```

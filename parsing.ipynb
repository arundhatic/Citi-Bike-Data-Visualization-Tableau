```python
import pandas as pd
import numpy as np
from geopy import distance
from geopy import Point
```

```python
rawData = pd.read_csv("data/nybike_from_url_2018_2019_2020.csv")
df_18_19_20 = pd.DataFrame(rawData)
df_18_19_20.head()
```

```python
df_18_19_20.tail()
```

```python
df_18_19_20['Gender']
```

```python
df_18_19_20['Gender'] = df_18_19_20['Gender'].replace([0,1,2],['unknown',"Male","Female"])
df_18_19_20["Distance"] = ''
df_18_19_20.tail()
```

```python
for index, row in df_18_19_20.iterrows():
    coords_1 = (row['Start Station Latitude'],row['Start Station Longitude'])
    coords_2 = (row['End Station Latitude'],row['End Station Longitude'])
    df_18_19_20.at[index,'Distance'] = distance.distance(coords_1, coords_2).km  
   # df_18_19_20.set_value(index,'distance',distance.vincenty(coords_1, coords_2).km)
df_18_19_20.tail()

```
